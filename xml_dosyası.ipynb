```python
import xml.etree.ElementTree as ET

def get_user_data():
    title = input("Başlık: ")
    subject = input("Anahtar Kelimeler: ")
    description = input("Açıklama: ")
    creator = input("Oluşturan kişi/kuruluş: ")
    license_ = input("Lisans: ")
    language = input("Dil: ")
    identifier = input("Kaynak URL'si: ")

    data = {
        'title': title,
        'subject': subject,
        'description': description,
        'creator': creator,
        'license': license_,
        'language': language,
        'identifier': identifier
    }
    return data

def write_to_xml(data):
    try:
        tree = ET.parse('/content/aek_data.xml')
        root = tree.getroot()
    except FileNotFoundError:
        root = ET.Element("records")

    record = ET.SubElement(root, "record")

    for key, value in data.items():
        element = ET.SubElement(record, key)
        element.text = value

    tree = ET.ElementTree(root)
    tree.write("/content/aek_data.xml")

    print("Veriler XML dosyasına kaydedildi!")

def read_from_xml():
    try:
        tree = ET.parse('/content/aek_data.xml')
        root = tree.getroot()

        for record in root.findall('record'):
            title = record.find('title').text
            subject = record.find('subject').text
            description = record.find('description').text
            creator = record.find('creator').text
            license_ = record.find('license').text
            language = record.find('language').text
            identifier = record.find('identifier').text

            print(f"Başlık: {title}")
            print(f"Anahtar Kelimeler: {subject}")
            print(f"Açıklama: {description}")
            print(f"Oluşturan: {creator}")
            print(f"Lisans: {license_}")
            print(f"Dil: {language}")
            print(f"URL: {identifier}")
            print("="*30)

    except FileNotFoundError:
        print("XML dosyası bulunamadı!")

def search_by_title(title_search):
    try:
        tree = ET.parse('/content/aek_data.xml')
        root = tree.getroot()
        found = False

        for record in root.findall('record'):
            title = record.find('title').text
            if title_search.lower() in title.lower():
                print(f"Başlık: {title}")
                print(f"Anahtar Kelimeler: {record.find('subject').text}")
                print(f"Açıklama: {record.find('description').text}")
                print(f"Oluşturan: {record.find('creator').text}")
                print(f"Lisans: {record.find('license').text}")
                print(f"Dil: {record.find('language').text}")
                print(f"URL: {record.find('identifier').text}")
                print("="*30)
                found = True

        if not found:
            print(f"{title_search} başlıklı kaynak bulunamadı.")

    except FileNotFoundError:
        print("XML dosyası bulunamadı!")

def main():
    print("Açık Eğitim Kaynağı (AEK) XML Yönetim Sistemi")
    data = None

    while True:
        print("\n1. Veri Gir (Başlık, Anahtar Kelimeler, vb.)")
        print("2. Veriyi XML Dosyasına Kaydet")
        print("3. XML Dosyasını Oku ve Listele")
        print("4. Başlıkla Arama Yap")
        print("5. Çıkış")

        choice = input("\nBir seçenek girin: ")

        if choice == '1':
            data = get_user_data()
            print("Veriler alındı.")

        elif choice == '2':
            if data:
                write_to_xml(data)
            else:
                print("Önce veri girmelisiniz.")

        elif choice == '3':
            read_from_xml()

        elif choice == '4':
            if data:
                search_term = input("Aramak istediğiniz başlık: ")
                search_by_title(search_term)
            else:
                print("Önce veri girmelisiniz.")

        elif choice == '5':
            print("Çıkılıyor...")
            break

        else:
            print("Geçersiz seçenek! Lütfen tekrar deneyin.")

main()
```
